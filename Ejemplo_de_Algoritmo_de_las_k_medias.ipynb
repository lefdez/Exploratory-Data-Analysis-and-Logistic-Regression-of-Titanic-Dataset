# Algoritmo de las k-medias

El algoritmo de k-medias (k-means) es un método de aprendizaje no supervisado utilizado para resolver problemas de clustering. Su objetivo es agrupar un conjunto de datos en kk grupos o clusters, donde cada grupo contiene puntos similares entre sí y diferentes de los de otros clusters.

El algoritmo comienza seleccionando aleatoriamente kk puntos como centroides iniciales. Luego sigue un proceso iterativo:

1.   Asignación: Cada punto de datos se asigna al cluster cuyo centroide esté más cerca, basado en una medida de distancia, como la euclidiana.
2.   Actualización: Los centroides de los clusters se recalculan como la media de los puntos asignados a cada uno.

Este proceso se repite hasta que los centroides no cambian significativamente, o hasta alcanzar un número máximo de iteraciones. El resultado es una partición de los datos en clusters definidos por sus centroides.

El algoritmo de k-medias es rápido y eficiente, especialmente en conjuntos de datos grandes. Sin embargo, tiene limitaciones: requiere especificar el número de clusters kk previamente, puede quedar atrapado en mínimos locales y es sensible a valores atípicos. Se utiliza ampliamente en aplicaciones como segmentación de clientes, compresión de imágenes y análisis de comportamiento. Es una herramienta esencial en análisis exploratorio de datos y minería de datos.

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
```

```python
datos = make_blobs(n_samples=200, n_features=2, centers=4)
```

```python
plt.scatter(datos[0][:,0], datos[0][:,1])
```

```python
modelo = KMeans(n_clusters=4)
modelo.fit(datos[0])
```

```python
modelo.cluster_centers_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.scatter(datos[0][:, 0], datos[0][:, 1], c=modelo.labels_, s=50, cmap='viridis')
ax1.set_title('Algoritmo de K-medias')

ax2.scatter(datos[0][:, 0], datos[0][:, 1], c=datos[1], s=50, cmap='viridis')
ax2.set_title('Datos observados')
```
